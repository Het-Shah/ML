```python
train=pd.read_csv('train.csv')
```

```python
X=train[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad',
       'tax', 'ptratio', 'black', 'lstat']]
y=train[['medv']]
```

```python
from sklearn import preprocessing
```

```python
X=preprocessing.scale(X)
y=preprocessing.scale(y)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=101)
```

```python
y_train=y_train.as_matrix()
```

```python
B=np.random.rand(13,1)
y_train.shape
```

```python
a=0.00001
con=1000
epoch=23
#_train.shape
```

```python
while epoch>=0 :
    for i in range(38):
        pred=np.matmul(X_train[i:i+7],B)
        sub=y_train[i:i+7]-pred
        sub=sub.reshape(1,7)
        loss=sub**2
        print(loss[0][6])
        temp=np.matmul(sub,X_train[i:i+7])
        temp=temp.reshape(13,1)
        B=B+(a*temp/233)
    epoch=epoch-1
```

```python
predicition=np.matmul(X_test,B)
```

```python
import sklearn.metrics as mt
print("RMSE: ",np.sqrt(mt.mean_squared_error(y_test,predicition)))
```
